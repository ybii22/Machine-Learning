```python
# --------------------
# Step 3: Prepare the data
# --------------------
def prepare_country_stats(oecd_bli, gdp_per_capita):
    oecd_bli = oecd_bli[oecd_bli["INEQUALITY"] == "TOT"]
    oecd_pivot = oecd_bli.pivot(index="Country", columns="Indicator", values="Value")

    gdp_per_capita.rename(columns={"2015": "GDP per capita"}, inplace=True)
    gdp_per_capita = gdp_per_capita[["Country", "GDP per capita"]]

    merged = pd.merge(oecd_pivot, gdp_per_capita, on="Country")
    merged.set_index("Country", inplace=True)
    return merged[["GDP per capita", "Life satisfaction"]]

country_stats = prepare_country_stats(oecd_bli, gdp_per_capita)

X = np.c_[country_stats["GDP per capita"]]
y = np.c_[country_stats["Life satisfaction"]]

# --------------------
# Step 4: Visualize the data
# --------------------
country_stats.plot(kind='scatter', x="GDP per capita", y='Life satisfaction')
plt.title("Life Satisfaction vs GDP per capita")
plt.grid(True)
plt.show()

# --------------------
# Step 5: Train linear regression model
# --------------------
model = sklearn.linear_model.LinearRegression()
model.fit(X, y)

# --------------------
# Step 6: Make a prediction (contoh: Cyprus GDP = 22587)
# --------------------
X_new = [[22587]]
prediction = model.predict(X_new)
print("Prediksi Life Satisfaction untuk GDP 22587:", prediction[0][0])
```

## **Penjelasan**

1. Machine Learning adalah cabang dari kecerdasan buatan yang memungkinkan komputer untuk belajar dari data, mengenali pola, dan membuat keputusan tanpa harus diprogram secara eksplisit. Komputer memperbaiki kinerjanya secara bertahap berdasarkan pengalaman yang didapat dari data.
2. dibawah ini:
* Klasifikasi (Classification): Menentukan kategori atau label dari suatu data baru berdasarkan data berlabel sebelumnya. Contoh: mendeteksi email spam vs. non-spam.

* Regresi (Regression): Memprediksi nilai numerik kontinu. Contoh: memprediksi harga rumah, memprediksi permintaan listrik.

* Clustering (Pengelompokan) / Segmentasi: Mengelompokkan data tanpa label ke dalam beberapa kelompok berdasar kemiripan. Contoh: segmentasi pelanggan untuk pemasaran.

* Deteksi Anomali (Anomaly Detection): Mencari data yang menyimpang jauh dari mayoritas atau pola umum. Contoh: mendeteksi transaksi kartu kredit yang mencurigakan, mendeteksi kerusakan pada mesin industri.

3. Labeled training set adalah himpunan data yang setiap contoh (instance)  baik itu vektor fitur, gambar, teks, dsb.  sudah dilengkapi label atau target yang benar.

4. dibawah ini:
* Klasifikasi: memprediksi kategori, misalnya email termasuk spam atau tidak.

* Regresi: memprediksi nilai numerik, seperti memperkirakan harga rumah berdasarkan luas dan lokasi.

5. dibawah ini:
* Clustering (pengelompokan): mengelompokkan data berdasarkan kesamaan.

* Reduksi dimensi: menyederhanakan data tanpa kehilangan informasi penting.

* Deteksi anomali: mendeteksi data yang menyimpang dari pola umum.

* Pencarian aturan asosiatif: menemukan pola hubungan antar item, seperti pada data belanja konsumen.

6. Algoritma yang cocok adalah Reinforcement Learning. Dalam metode ini, robot belajar melalui percobaan dan kesalahan, dan mendapatkan umpan balik dari lingkungan untuk memperbaiki tindakannya di masa depan.

7. Algoritma yang digunakan adalah unsupervised learning, khususnya metode clustering. Teknik ini digunakan untuk mengelompokkan pelanggan berdasarkan perilaku atau karakteristik yang serupa tanpa adanya label sebelumnya.

8. Masalah deteksi spam termasuk dalam supervised learning, karena model dilatih menggunakan email yang sudah diberi label "spam" dan "bukan spam", sehingga bisa belajar membedakan keduanya dari pola data.

9. Online learning system adalah sistem pembelajaran mesin yang menerima data satu per satu atau dalam aliran terus-menerus, dan langsung memperbarui model tanpa harus melatih ulang dari awal. Cocok untuk data besar atau sistem real-time seperti rekomendasi konten atau deteksi penipuan.

10. Out-of-core learning adalah teknik untuk melatih model pada data yang terlalu besar untuk dimuat sekaligus ke dalam memori. Data diproses secara bertahap dalam batch kecil menggunakan algoritma yang mendukung pembelajaran bertahap (incremental).

11. Algoritma instance-based learning seperti K-Nearest Neighbors (KNN) menggunakan ukuran kemiripan (misalnya jarak Euclidean) antara contoh baru dan contoh pelatihan untuk membuat prediksi.

12. dibawah ini:
* Parameter model adalah nilai yang dipelajari dari data selama pelatihan (contohnya bobot dalam regresi linier).

* Hyperparameter adalah nilai yang ditentukan sebelum pelatihan dan mengatur proses pembelajaran (contohnya jumlah tetangga di KNN, learning rate di neural network).

13. Algoritma pembelajaran berbasis model mencari parameter terbaik untuk model matematis yang memetakan input ke output. Strategi paling umum adalah meminimalkan fungsi biaya menggunakan teknik optimisasi seperti gradient descent. Prediksi dibuat dengan memasukkan data baru ke dalam model yang telah dilatih.

14. empat tantangan utama:
*  Kekurangan data pelatihan berkualitas
* Overfitting (model terlalu menyesuaikan data pelatihan)
* Underfitting (model terlalu sederhana)
* Data bias atau tidak representatif
Tantangan lain bisa mencakup data noise, fitur yang tidak relevan, dan komputasi yang mahal.

15. tiga solusi:
* Gunakan model yang lebih sederhana.
* Tambah data pelatihan.
* Gunakan regularisasi untuk membatasi kompleksitas model.

16. Test set adalah data yang tidak digunakan saat pelatihan, digunakan untuk mengukur kemampuan model dalam menggeneralisasi pada data baru. Ini penting agar evaluasi performa model menjadi objektif dan tidak bias.

17. Validation set digunakan untuk menyetel hyperparameter dan memilih model terbaik. Validation membantu mengevaluasi model selama pelatihan tanpa menyentuh data uji (test set).

18. Train-dev set adalah subset dari data pelatihan yang digunakan seperti validation set ketika data uji sangat berbeda dari pelatihan. Digunakan untuk memastikan apakah masalah ada pada overfitting atau distribusi data. Biasanya digunakan jika performa pada validation set sangat buruk.

19. Jika test set digunakan untuk menyetel hyperparameter, maka test set tidak lagi menjadi tolok ukur yang objektif. Ini bisa menyebabkan data leakage dan hasil evaluasi menjadi terlalu optimis, sehingga performa model di dunia nyata bisa jauh lebih buruk.
