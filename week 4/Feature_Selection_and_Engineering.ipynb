## **1. Membuat Data**

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split

# Membuat dataset contoh untuk klasifikasi
X, y = make_classification(n_samples=1000, n_features=20, n_classes=2, random_state=42)

# Membagi dataset menjadi training dan testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

```

## **2. Membuat Model Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression

# Membuat model Logistic Regression
model = LogisticRegression(max_iter=1000)

# Melatih model
model.fit(X_train, y_train)
```

## **3. Evaluasi Model**

Akurais, Presisi, Recall, F1 Score, AUC, dan ROC

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, roc_curve
import matplotlib.pyplot as plt

# Prediksi pada data uji
y_pred = model.predict(X_test)
y_prob = model.predict_proba(X_test)[:, 1]  # probabilitas untuk kelas positif

# Evaluasi menggunakan berbagai metrik
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
auc = roc_auc_score(y_test, y_prob)

# Menghitung ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_prob)

# Menampilkan hasil evaluasi
print(f"Akurasi: {accuracy:.4f}")
print(f"Presisi: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
print(f"AUC: {auc:.4f}")

# Plot ROC curve
plt.plot(fpr, tpr, label='ROC Curve (AUC = {:.4f})'.format(auc))
plt.plot([0, 1], [0, 1], linestyle='--', label='Random Classifier')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc='lower right')
plt.show()
```

## **4. Penjelasan Metrik Evaluasi**
1. Akurasi : mengukur proporsi prediksi yang benar.

Akurasi= TP+TN/TP+TN+FP+FN

dimana TP = True Positives, TN = True Negatives, FP = False Positives, FN = False Negatives.

2. Presisi: Mengukur proporsi positif yang benar

Presisi= TP/TP+FP

3. Recal : mengukur proporsi data positif yang terdeteksi dengan benar.

Recall= TP/TP+FN

4. F1 Score : Harmonik rata-rata antara presisi dan recall.


F1=2× Presisi×Recall/Presisi+Recall

5. AUC (Area Under the Curve): Mengukur area di bawah kurva ROC. Semakin tinggi AUC, semakin baik model dalam membedakan kelas positif dan negatif.

6. ROC Curve: Grafik yang menunjukkan hubungan antara true positive rate (recall) dan false positive rate pada berbagai threshold.




## **5. Mengevaluasi Akurasi Model**

```python
# Evaluasi pada training set
y_train_pred = model.predict(X_train)
train_accuracy = accuracy_score(y_train, y_train_pred)
print(f"Akurasi pada Training Set: {train_accuracy:.4f}")

# Evaluasi pada testing set
test_accuracy = accuracy
print(f"Akurasi pada Testing Set: {test_accuracy:.4f}")
```
